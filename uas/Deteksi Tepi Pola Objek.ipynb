```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("foto.jpeg")
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
# Ubah foto menjadi hitam putih
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (15, 15), 0)

edges = cv.Canny(blurred, 38, 80)

contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

contour_image = image_rgb.copy()

cv.drawContours(contour_image, contours, -1, (0, 255, 0), 2)
```

```python
print("Output 1") 
fig1, axs1 = plt.subplots(1, 2, figsize=(10, 5))

axs1[0].imshow(image_rgb)
axs1[0].set_title('Original Image')

axs1[1].imshow(edges, cmap="gray")
axs1[1].set_title('Canny Edge Detection')

plt.tight_layout()
plt.show() # Tampilkan gambar bagian atas


print("Output 2")
fig2, axs2 = plt.subplots(1, 3, figsize=(15, 5))

axs2[0].imshow(image_rgb)
axs2[0].set_title('Original Image')

axs2[1].imshow(edges, cmap="gray")
axs2[1].set_title('Canny Edge Detection')

axs2[2].imshow(contour_image)
axs2[2].set_title('Contours Detection')

plt.tight_layout()
plt.show()
```
